# Filtering 

```python
# How Many People Take Cappuccino in Morning???

m1=df['coffee_name']=='Cappuccino'
m2=df['Time_of_Day']=='Morning'

df[m1 & m2].shape[0]
```

```python

# How Many People Take 'Latte' Coffee in AfterNoon ???

m1=df['coffee_name']=='Latte'
m2=df['Time_of_Day']=='Afternoon'

df[m1 & m2].shape[0]
```

```python
# which Type of coffee are high demand  sale in morning ??

fil=df[(df['Time_of_Day']=='Morning') & (df['coffee_name'])]
fil[['Time_of_Day','coffee_name']].value_counts().reset_index(name='Total Coffee Sales')





```

```python
# which Type of coffee are high demand  sale in Night ??

fil=df[(df['Time_of_Day']=='Night') & (df['coffee_name'])]
fil[['Time_of_Day','coffee_name']].value_counts().reset_index(name='Total Coffee Sales')
```

```python
# how many people Latte coffee take on sunday night??


mak1=df['Time_of_Day']=='Night'
mak2=df['coffee_name']=='Latte'
mak3=df['Weekday']=='Sun'

df[mak1 & mak2 & mak3].shape[0]
```

```python
# Analyse the coffee Type in Time of day.

m2=df['Time_of_Day']
m1=df['coffee_name']

fil=df.groupby([m2 ,m1]).size().reset_index(name='Total Coffee Sale')



fil.value_counts().reset_index()
```
